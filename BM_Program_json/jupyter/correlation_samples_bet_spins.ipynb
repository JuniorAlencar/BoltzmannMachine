```python
sampleN1 = "004"
path_1 = f'../Results/Flow_Spin/Data_01/iRB_064x128_M1.0e+00_u1.0e-02_U1.0e-02_Ca1.00e-01/S0.50_n4711/Configs/c{sampleN1}_Satu.dat'
df_1 = pd.read_csv(path_1, header=None, delimiter=',')
num_columns = df_1.shape[1]
headers = [f's_{i+1}' for i in range(num_columns)]
df_1.columns = headers
df_1.head()
```

```python
sampleN2 = "053"
path_2 = f'../Results/Flow_Spin/Data_01/iRB_064x128_M1.0e+00_u1.0e-02_U1.0e-02_Ca1.00e-01/S0.50_n4711/Configs/c{sampleN2}_Satu.dat'
df_2 = pd.read_csv(path_2, header=None, delimiter=',')
num_columns = df_2.shape[1]
headers = [f's_{i+1}' for i in range(num_columns)]
df_2.columns = headers
df_2.head()
```

```python
df_1_means = df_1.mean()
df_2_means = df_2.mean()

hist_1, bin_edges_1 = np.histogram(df_1_means, bins=20)
# Normalizar o histograma para que seja uma distribuição de probabilidade
P_k_1 = hist_1 / np.sum(hist_1)

# Calcular o centro de cada bin (os valores de k)
k_1 = (bin_edges_1[:-1] + bin_edges_1[1:]) / 2

hist_2, bin_edges_2 = np.histogram(df_2_means, bins=20)
# Normalizar o histograma para que seja uma distribuição de probabilidade
P_k_2 = hist_2 / np.sum(hist_2)

# Calcular o centro de cada bin (os valores de k)
k_2 = (bin_edges_2[:-1] + bin_edges_2[1:]) / 2


plt.plot(k_1, P_k_1, marker='o', linestyle='-', markersize=15, mec='k', mfc='none', color='k', label='sample_1_spin1')
plt.plot(k_2, P_k_2, marker='o', linestyle='-', markersize=15, mec='r', mfc='none', color='r', label='sample_2_spin1')
plt.legend()
plt.show()
```
