```python
props = ["Cij-ising", "mi-ising", "Pij-ising", "sisj-ising", "sisjsk-ising", "Tijk-ising"]
method = ["exact", "metropolis"]
file_sample = "sampleN20"
dict_values = {"Cij-ising-exact":[],"mi-ising-exact":[],"Pij-ising-exact":[], "sisj-ising-exact":[], "sisjsk-ising-exact":[], "Tijk-ising-exact":[],
               "Cij-ising-metropolis":[],"mi-ising-metropolis":[],"Pij-ising-metropolis":[], "sisj-ising-metropolis":[], "sisjsk-ising-metropolis":[], "Tijk-ising-metropolis":[]}

for prop in props:
    x = [] # exact
    y = [] # metropolis
    for m in method:
        path_patterns = [f"../Results_{m}/SeparateData/{prop}"]

        for comp in path_patterns:
            if(prop == "Cij-ising"):
                file_patern = "Cij_ising"
            elif(prop=="mi-ising"):
                file_patern = "mi_ising"
            elif(prop=="Pij-ising"):
                file_patern = "Pij_ising"
            elif(prop=="sisj-ising"):
                file_patern = "sisj_ising"
            
            elif(prop=="sisjsk-ising"):
                file_patern = "sisjsk_ising"
            
            elif(prop == "Tijk-ising"):
                file_patern = "Tijk_ising"
            if(m == "exact"):
                err_j, err_h, mteq, mrelx = 2.00e-05, 2.00e-05, 150, 2
                filename = f"{comp}/{file_patern}_{file_sample}_err_j_{err_j:.2e}_err_h_{err_h:.2e}_mteq_{mteq}_mrelx_{mrelx}.dat"
                with open(filename, "r") as fileE:
                    dict_values[f"{prop}-exact"] = [float(line.strip()) for line in fileE]
            else:
                err_j, err_h, mteq, mrelx = 1.00e-03, 5.87e-04, 150, 2
                filename = f"{comp}/{file_patern}_{file_sample}_err_j_{err_j:.2e}_err_h_{err_h:.2e}_mteq_{mteq}_mrelx_{mrelx}.dat"
                with open(filename, "r") as fileM:
                    dict_values[f"{prop}-metropolis"] = [float(line.strip()) for line in fileM]
```

```python
for prop in props:
    plt.plot(dict_values[f"{prop}-exact"],dict_values[f"{prop}-metropolis"],'o', label=f'{prop}')
    plt.plot(dict_values[f"{prop}-exact"],dict_values[f"{prop}-exact"])
    plt.xlabel("exact")
    plt.ylabel("metropolis")
    plt.legend()
    plt.show()
```
