```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
import sys
```

```python
import json

f = open('../Results_Metropolis/sampleN20_j_min_1.00e-04_h_min_1.00e-04_t_eq_3000.json')
f2 = open('../Results_Metropolis/sampleN20_j_min_7.05e-04_h_min_5.80e-04_t_eq_3000.json')
f3 = open('../Results_Metropolis/sampleN20_j_min_9.40e-04_h_min_5.94e-04_t_eq_3000.json')

# returns JSON object as 
# a dictionary
data = json.load(f)
data2 = json.load(f2)
data3 = json.load(f3)
# Print all keys in the JSON dictionary
for key in data.keys():
    print(key)
```

```python
prop = "SSS"
S_ising = data[f'{prop}_exp']
S_exp = data[f'{prop}_ising']

S_ising_2 = data2[f'{prop}_exp']
S_exp_2 = data2[f'{prop}_ising']

S_ising_3 = data3[f'{prop}_exp']
S_exp_3 = data3[f'{prop}_ising']

plt.figure(figsize=(16,7))
plt.plot(S_exp,S_ising,'o',ms=15,color='b')
plt.plot(S_exp_2,S_ising_2,'o',ms=15,color='r')
plt.plot(S_exp_3,S_ising_3,'o',ms=15,color='g')

plt.plot(S_exp, S_exp,'--',color='b')
plt.plot(S_exp_2, S_exp_2,'--',color='r')
plt.plot(S_exp_3, S_exp_3,'--',color='g')

plt.xlabel("exp",fontsize=20)
plt.ylabel("ising",fontsize=20)
plt.title(prop,fontsize=20)
plt.show()
```
